```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# New section

FOR STRAIGHT LINE AND CORRIDOR PATH


```
fileName = "straightLine.csv"
df = pd.read_csv("/straightLine.csv")
index = range(1, len(df['Gx']) + 1)

plt.rcParams["figure.figsize"] = (12,6)
fig = plt.figure()
fig.suptitle("Straight path Gyroscope Accelerometer and Magnatometer data")
plt.plot(index, df['Gx'], 'g.', label='x', linestyle='solid', marker=',')
plt.plot(index, df['Gy'], 'b.', label='y', linestyle='solid', marker=',')
plt.plot(index, df['Gz'], 'r.', label='z', linestyle='solid', marker=',')
plt.title("Gyroscope")
plt.xlabel("Sample #")
plt.ylabel("Gyroscope(G)")
plt.legend()
plt.show()

plt.plot(index, df['Ax'], 'g.', label='x', linestyle='solid', marker=',')
plt.plot(index, df['Ay'], 'b.', label='y', linestyle='solid', marker=',')
plt.plot(index, df['Az'], 'r.', label='z', linestyle='solid', marker=',')
plt.title("Acceleration")
plt.xlabel("Sample #")
plt.ylabel("Acceleration (G)")
plt.legend()
plt.show()

plt.plot(index, df['Mx'], 'g.', label='x', linestyle='solid', marker=',')
plt.plot(index, df['My'], 'b.', label='y', linestyle='solid', marker=',')
plt.plot(index, df['Mz'], 'r.', label='z', linestyle='solid', marker=',')
plt.title("Magnetometer")
plt.xlabel("Sample #")
plt.ylabel("Magnetometer (G)")
plt.legend()
plt.show()
```
